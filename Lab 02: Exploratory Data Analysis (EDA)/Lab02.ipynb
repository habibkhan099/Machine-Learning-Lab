<font color='aqua'>• Plot histograms, boxplots, and KDE plots to understand distributions and detect 
outliers.</font>

```python
# Plotting histograms
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
sns.histplot(df['Age'].dropna(), kde=True, bins=20, color='blue')
plt.title('Age Distribution')

plt.subplot(1,3,2)
sns.histplot(df['Fare'].dropna(), kde=True, bins=20, color='green' )
plt.title('Fare Distribution')

plt.subplot(1,3,3)
sns.histplot(df['Parch'].dropna(), kde=True, bins=20, color='red')
plt.title("Parch Distribution")

```

```python
# Boxplot
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
sns.boxplot(df['Age'].dropna(), color='blue')
plt.title('Boxplot of Age')

plt.subplot(1, 3, 2)
sns.boxplot(df['Fare'].dropna(), color='green')
plt.title('Boxplot of Fare')

plt.subplot(1, 3, 3)
sns.boxplot(df['Parch'].dropna(), color='red')
plt.title('Boxplot of Parch')
plt.tight_layout()
plt.show()
```

```python
# Plotting KDE (Kernel Density Estimation) plots
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
sns.kdeplot(df['Age'].dropna(), color='blue', shade=True)
plt.title('KDE of Age')

plt.subplot(1, 3, 2)
sns.kdeplot(df['Fare'].dropna(), color='green', shade=True)
plt.title('KDE of Fare')

plt.subplot(1, 3, 3)
sns.kdeplot(df['Parch'].dropna(), color='red', shade=True)
plt.title('KDE of Parch')

plt.tight_layout()
plt.show()
```

## <font color='aqua'>Task 3: Categorical Data Analysis
### 3. Question: What is the frequency distribution of categorical columns?
• Explore Pclass, Sex, Embarked, and Survived using value_counts().
</font>

## Check Frequency Distribution Using value_counts()

```python
# Frequency distribution of categorical columns
print("Pclass distribution:")
print(df['Pclass'].value_counts())

print("\nSex distribution:")
print(df['Sex'].value_counts())

print("\nEmbarked distribution:")
print(df['Embarked'].value_counts())

print("\nSurvived distribution:")
print(df['Survived'].value_counts())
```

<font color='aqua'>. Visualize the distributions using countplots and pie charts.</font>

## Visualization using countplots()

```python
# Visualizing categorical columns using countplot
plt.figure(figsize=(12, 6))

# Pclass distribution
plt.subplot(2, 2, 1)
sns.countplot(x='Pclass', data=df, palette='Set2')
plt.title('Pclass Distribution')

# Sex distribution
plt.subplot(2, 2, 2)
sns.countplot(x='Sex', data=df, palette='Set2')
plt.title('Sex Distribution')

# Embarked distribution
plt.subplot(2, 2, 3)
sns.countplot(x='Embarked', data=df, palette='Set2')
plt.title('Embarked Distribution')

# Survived distribution
plt.subplot(2, 2, 4)
sns.countplot(x='Survived', data=df, palette='Set2')
plt.title('Survived Distribution')

plt.tight_layout()
plt.show()
```

## Visualization using pie charts

<font color='aqua'></font>

```python
# Visualizing categorical columns using pie charts
plt.figure(figsize=(12, 6))

# Pclass distribution pie chart
plt.subplot(2, 2, 1)
df['Pclass'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff','#99ff99'])
plt.title('Pclass Distribution')
plt.ylabel('')

# Sex distribution pie chart
plt.subplot(2, 2, 2)
df['Sex'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff'])
plt.title('Sex Distribution')
plt.ylabel('')

# Embarked distribution pie chart
plt.subplot(2, 2, 3)
df['Embarked'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff','#99ff99'])
plt.title('Embarked Distribution')
plt.ylabel('')

# Survived distribution pie chart
plt.subplot(2, 2, 4)
df['Survived'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff'])
plt.title('Survived Distribution')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

## <font color='aqua'>Task 4: Relationship Between Variables (Bivariate Analysis)
### 4. Question: How are variables related to each other?
### Analyze the relationship between:
### ▪ Numerical-Numerical: 
Age vs Fare (scatter plot and correlation).</font>

```python
# Scatter plot of Age vs Fare
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Fare', data=df, color='blue')
plt.title('Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()

# Correlation between Age and Fare
correlation = df[['Age', 'Fare']].corr()
print(f"Correlation between Age and Fare: \n{correlation}")
```

<font color='aqua'>▪ Categorical-Categorical: Pclass vs Survived (grouped bar chart).</font>

```python
# Grouped bar chart of Pclass vs Survived
plt.figure(figsize=(8, 6))
sns.countplot(x='Pclass', hue='Survived', data=df, palette='Set2')
plt.title('Pclass vs Survived')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.show()
```

<font color='aqua'>▪ Numerical-Categorical: Age distribution across survival status (boxplot).</font>

```python
# Boxplot of Age distribution across Survived status
plt.figure(figsize=(8, 6))
sns.boxplot(x='Survived', y='Age', data=df, palette='Set2')
plt.title('Age distribution across Survived status')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Age')
plt.show()
```

## <font color='aqua'>Task 5: Multivariate Analysis
### 5. Question: How do multiple variables interact with each other?
• Perform a pairwise analysis on selected columns such as Age, Fare, Pclass, and 
Survived using pairplots.
</font>

```python
# Pairplot of selected columns (Age, Fare, Pclass, Survived)
sns.pairplot(df[['Age', 'Fare', 'Pclass', 'Survived']], hue='Survived', palette='Set1')
plt.suptitle('Pairwise Analysis: Age, Fare, Pclass, Survived', y=1.02)

```

<font color='aqua'>• Visualize overall correlations using a heatmap.</font>

```python
# Compute the correlation matrix
correlation_matrix = df[['Age', 'Fare', 'Pclass', 'Survived']].corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=1, linecolor='black')
plt.title('Correlation Heatmap: Age, Fare, Pclass, Survived')
plt.show()
```
